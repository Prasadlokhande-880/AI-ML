```python
import numpy as np
import pandas as py
```

```python
## this is the code for the importing the data set
data=py.read_csv('./Car_Purchasing_Data.csv');
print(data);
```

```python
import matplotlib.pyplot as plt

annual_salary = data['Annual Salary']
car_purchase_amount = data['Car Purchase Amount']

# Plotting the scatter plot
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
plt.scatter(annual_salary, car_purchase_amount, color='blue', alpha=0.5)  # Adjust color and transparency as needed
plt.title('Annual Salary vs Car Purchase Amount')
plt.xlabel('Annual Salary')
plt.ylabel('Car Purchase Amount')
plt.grid(True)  # Optional: Add gridlines
plt.show()
```

```python
## storing the data frams value's into theothere var

X=data['Annual Salary']
Y=data["Car Purchase Amount"]

print(X,Y)
```

```python
avg_X=np.average(X)
avg_Y=np.average(Y)

print(avg_X,avg_Y)
```

```python
## this is the coede for calculating the X-avg_X

sub_X=X-avg_X
sub_Y=Y-avg_Y

print(sub_X,sub_Y)
```

```python
import numpy as np

# Assuming sub_X and sub_Y are your variables
pro = np.multiply(sub_X, sub_Y)  # Element-wise multiplication
sqr = np.power(sub_X, 2)  # Squaring sub_X element-wise

sum1 = np.sum(pro)
sum2 = np.sum(sqr)
slope = np.divide(sum1, sum2)

print(slope)
```

```python
## this the code for the claculating const

const=-avg_X*slope+avg_Y

print(const)
```

```python
import matplotlib.pyplot as plt

# Assuming you have data points in sub_X and sub_Y
# Plotting the scatter plot of data points
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
plt.scatter(sub_X, sub_Y, color='blue', label='Data Points')  # Plotting the data points

# Calculating the predicted values based on the regression line equation (y = mx + c)
predicted_Y = slope * sub_X + const

# Plotting the regression line
plt.plot(sub_X, predicted_Y, color='red', label='Regression Line')  # Plotting the regression line

plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()  # Show legend
plt.grid(True)  # Optional: Add gridlines
plt.show()
```
